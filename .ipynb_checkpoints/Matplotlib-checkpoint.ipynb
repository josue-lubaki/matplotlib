```python
# LINE CHART
```

```python
plt.figure()
x = np.arange(10)
y = np.square(x)

# Add color (c) ; linestyle (ls) et linewidth (lw)
plt.plot(x,y, c="red", ls="--", lw=3, label="Y=X^2")

# Ajout des labels et de fontsize
plt.xlabel("X Axis", fontsize=14)
plt.ylabel("Y Axis", fontsize=14)

# Modifier les écarts des valeurs sur l'axe X et Y
plt.xticks(np.arange(0,10, step=1))
plt.yticks(np.arange(0,100,step=5))

# Ajout du title
plt.title("First Line Chart", fontsize=16)

# Ajout du grid
plt.grid()

# Ajout de la legende en ajoutant le label au plt.plot()
plt.legend(loc="upper center")

# Affichage
plt.show()

# Sauvergarder
# plt.savefig("line_chart.png")
```

```python
plt.figure()
x = np.arange(10)
y = np.square(x)
y2 = np.power(x,3)

# Add color (c) ; linestyle (ls) et linewidth (lw)
plt.plot(x,y, c="red", ls="--", lw=3, label="Y=X^2")
plt.plot(x,y2, ls="--", lw=3, label="Y=X^3")

# Ajout des labels et de fontsize
plt.xlabel("X Axis", fontsize=14)
plt.ylabel("Y Axis", fontsize=14)

# Modifier les écarts des valeurs sur l'axe X et Y
plt.xticks(np.arange(0,10, step=1))
# plt.yticks(np.arange(0,100,step=5))

# Ajout du title
plt.title("First Line Chart", fontsize=16)

# Ajout du grid
plt.grid()

# Ajout de la legende en ajoutant le label au plt.plot()
plt.legend(loc="upper center")

# Affichage
plt.show()

# Sauvergarder
# plt.savefig("line_chart.png")
```

```python
plt.figure(figsize=(10,6))

x = np.arange(10)
y = np.square(x)
y2 = np.power(x,3)

# subplot(@nbreRows, @nbreCols, @position)
plt.subplot(2,1,1)
plt.plot(x,y, c="red", ls="--", lw=3, label="y=x^2")
plt.ylabel("Y Axis", fontsize=14)
plt.xticks(np.arange(0,10, step=1))
plt.title("First Line Chart", fontsize=16)
plt.grid()
plt.legend(loc="upper center")

# subplot(@nbreRows, @nbreCols, @position)
plt.subplot(2,1,2)
plt.plot(x,y2, label="y=x^3")
plt.xlabel("X Axis", fontsize=14)
plt.ylabel("Y Axis", fontsize=14)
plt.xticks(np.arange(0,10, step=1))
plt.grid()
plt.legend(loc="upper center")

plt.show()
```

```python
# LINE CHART (Array)
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20,15))

x1 = np.arange(10)
y1 = np.square(x1)
y2 = np.power(x1,3)
x2 = np.arange(50)

y3 = np.sin(x2)
y4 = np.cos(x2)

ax[0,0].plot(x1,y1, label="y=x^2", c="red")
ax[0,0].set_xlabel("X-Axis", fontsize=14)
ax[0,0].set_ylabel("Y-Axis", fontsize=14)
ax[0,0].set_title("First plot")
ax[0,0].set_xticks(np.arange(0,10, step=1))
ax[0,0].legend()
ax[0,0].grid()

ax[1,0].plot(x1,y2, label="y=x^3", c="green")
ax[1,0].set_xlabel("X-Axis", fontsize=14)
ax[1,0].set_ylabel("Y-Axis", fontsize=14)
ax[1,0].set_title("Second plot")
ax[1,0].set_xticks(np.arange(0,10, step=1))
ax[1,0].legend()
ax[1,0].grid()

ax[0,1].plot(x2,y3, label="y=sin(x2)", c="darkblue")
ax[0,1].set_xlabel("X-Axis", fontsize=14)
ax[0,1].set_ylabel("Y-Axis", fontsize=14)
ax[0,1].set_title("Third plot")
ax[0,1].legend()

ax[1,1].plot(x2,y4, label="y=cos(x2)")
ax[1,1].set_xlabel("X-Axis", fontsize=14)
ax[1,1].set_ylabel("Y-Axis", fontsize=14)
ax[1,1].set_title("Fourth plot")
ax[1,1].legend(loc="upper right")
```

```python
# AREA CHART
```

```python
plt.figure(figsize=(8,5))
x = np.arange(10)
y = np.square(x)
x2 = np.arange(50)
y3 = np.sin(x2)

plt.plot(x,y, c="red")

# Ajouter le remplissage entre x et y
plt.fill_between(x,y)

plt.xlabel("X Axis", fontsize=14)
plt.ylabel("Y Axis", fontsize=14)

plt.title("Area Chart", fontsize=16)
plt.grid()

# Affichage
plt.show()
```

```python
# HISTOGRAM
```

```python
df = pd.read_csv("movies3.csv")
```

```python
df.head()
```

```python
plt.figure()
plt.hist(df["rating"], bins=6, edgecolor="black", color="yellow")
plt.title("Histogram Example")
plt.xlabel("Rating", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks(np.arange(1,5, step=0.5))
plt.show()
```

```python
# BAR PLOT
```

```python
x = np.arange(10)
bar = np.square(x)
```

```python
plt.figure(figsize=(8,5))
plt.bar(x, bar, color="red", width=0.5, label="Red Bars")

# Ajouter les valeurs
for x_coord, y_coord in zip(x,bar):
    plt.text(x_coord, y_coord+1, y_coord, ha="center")

plt.title("Bar Chart")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,10))

plt.barh(x, bar)

plt.show()
```

```python
# SCATTER PLOT
```

```python
plt.figure()

plt.scatter(df["rating"], df["reviews"], s=200, c=df["like"])

plt.title("Scatter plot example")
plt.xlabel("Rating", fontsize=14)
plt.ylabel("Reviews", fontsize=14)
plt.show()
```

```python
df.head()
```

```python
df["like"].value_counts()
```

```python
# example using alpha option
for i in range(1,7):
    plt.figure(i)
    plt.scatter(df["rating"], df["reviews"], s=200, c=df["like"], alpha=0.2 * i)
    plt.title("Scatter plot example")
    plt.xlabel("Rating", fontsize=14)
    plt.ylabel("Reviews", fontsize=14)
    plt.show()
```

```python
# PIE CHART AND DONUT CHART
```

```python
plt.figure(figsize=(5,5))

segments = [15,25,10,25,5]
labels = ["segments_1", "segments_2","segments_3","segments_4","segments_5"]

plt.pie(segments, labels=labels, startangle = 45, autopct="%1.0f%%")

plt.show()
```

```python
plt.figure(figsize=(5,5))

segments = [15,25,10,25,5]
labels = ["segments_1", "segments_2","segments_3","segments_4","segments_5"]

plt.pie(segments, labels=labels, startangle = 45, autopct="%1.0f%%")

# Create circle
circle = plt.Circle((0,0), 0.7, color="white")

# Récupèrer la figure courante via gcf() function
get_current_figure = plt.gcf()

# Ajouter au figure courant le cercle
# gca() : getCurrentAccess function
get_current_figure.gca().add_artist(circle)

plt.show()
```
